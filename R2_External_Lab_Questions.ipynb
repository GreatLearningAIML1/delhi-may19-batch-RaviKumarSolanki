```
#Question 4
data["GoalsFor"]=data["GoalsFor"].astype("int64")
data["GoalsAgainst"]=data["GoalsAgainst"].astype("int64")
```

```
def Goal_diff_count(T,F,A):
  for row in np.arange(len(T)):
    gd=F-A
  return T, gd

```

```
team4,goal=Goal_diff_count(data["Team"],data["GoalsFor"],data["GoalsAgainst"])
```

```
Team4=pd.Series(data=goal.values,index=team4)
Team4
```

```
#Finding team with max Goal differencce
Team4.idxmax()
```

```
#Finding team with min Goal differencce
Team4.idxmin()
```
