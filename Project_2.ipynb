```
import numpy as np
import pandas as pd
from scipy import stats 
import matplotlib.pyplot as plt
%matplotlib inline
```

```
old=np.array([57,103,59,75,84,73,35,110,44,82,67,64,78,53,41,39,80,87,73,65,28,62,49,84,63,77,67,101,91,50])
new=np.array([62,122,54,82,84,86,32,104,38,107,84,85,99,39,34,58,73,53,66,78,41,71,38,95,81,58,75,94,100,68])
```

```
#Q1
Mean_old=np.mean(old)
Mean_new=np.mean(new)
```

```
print('Mean of old scheme is %f'%(Mean_old))
print('Mean of new scheme is %f'%(Mean_new))
```

```
#Q2
A=0.05
t_statistic,p_value=stats.ttest_ind(old,new)
print(t_statistic,p_value)
```

```
#Q3
def result(Pval,alpha):
  if Pval > alpha :
    print('Null hypothesis is accepted as Pvalue %f is greater than level of significance %.3f'%(Pval,alpha))
  else:
    print('Null hypothesis is rejected as Pvalue %f is less than level of significance %.3f'%(Pval,alpha))
result(p_value,A)
```
